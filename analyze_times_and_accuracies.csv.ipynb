```python
import datetime as dt
import pylab as plt
import pandas as pd
import numpy as np
```

```python
results_dir = "analyzed/"
today = dt.datetime.today().strftime('%m_%d_%Y')
```

```python
times = pd.read_csv( results_dir + today + "_times.csv" , header=None ).as_matrix()
accuracies = pd.read_csv( results_dir + today + "_accuracies.csv" , header=None ).as_matrix()
```

```python
for i in range( len( times[ 0 ] ) ):
    idx = np.where( accuracies[ i ] == 1 )[ 0 ]
    plt.scatter( [ i ] * len( idx ) , times[ i , idx ] )
    mean = plt.scatter( i , np.mean( times[ i , idx ] ) , color='k' , marker='x' )

plt.xlabel( "Model Number" )
plt.ylabel( "Time to Answer Question" )
plt.title( "Question Times by Model" )

plt.legend( [ mean ] , [ "Mean Response Time" ] )

plt.savefig( "question_times_by_model_"+today+".png" , bbox_inches="tight" )
plt.show()
```
